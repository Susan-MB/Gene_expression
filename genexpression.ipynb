Differential Expression Analysis - to find significantly up regulated and downregulated genes.

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind
```

```python
sample_info = pd.DataFrame(index=expr.columns)

sample_info["type"] = sample_info.index.to_series().apply(
    lambda x: "Tumor" if x.split("-")[3][:2] == "01" else "Normal"
)

sample_info["type"].value_counts()
```

```python
log_expr = np.log2(expr + 1)
```

```python
tumor_samples = sample_info[sample_info["type"] == "Tumor"].index
normal_samples = sample_info[sample_info["type"] == "Normal"].index
```

```python
tumor_expr = log_expr[tumor_samples]
normal_expr = log_expr[normal_samples]
```

```python
results = []
for gene in log_expr.index:
    tumor_values = tumor_expr.loc[gene]
    normal_values = normal_expr.loc[gene]
    stat, pval = ttest_ind(tumor_values, normal_values, nan_policy='omit')
    logFC = tumor_values.mean() - normal_values.mean()
    results.append([gene, logFC, pval])

```

```python
dea_expr = pd.DataFrame(results, columns=["gene", "logFC", "pvalue"])
```

Multiple testing correction - because thousands of genes were tested simultaneously, the equationw as applied ot control false discovery rate.

```python
from statsmodels.stats.multitest import multipletests

dea_expr["adj_pvalue"] = multipletests(dea_expr["pvalue"], method="fdr_bh")[1]
```

Identification of significantly expressed genes
- adjusted p_value < 0.05 and a biological threshold |logFC| > 1 to identify meaningful gene expression changes.

Volcano Plot


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    dea_expr["logFC"],
    -np.log10(dea_expr["pvalue"]),
    s=2
)

plt.axhline(-np.log10(0.05), color='red')
plt.axvline(-1, color='black')
plt.axvline(1, color='black')

plt.xlabel("log2 Fold Change")
plt.ylabel("-log10 p-value")
plt.title("Volcano Plot: Tumor vs Normal")
plt.show()
```

A volcano plot visualises differentiale expression results by plotting log2 fold change against statistical significance, -log10 p-value which allowed identification of genes that are both highly differentially expressed and statistically significant.

Top genes and the biological itnerpretation 


```python
sample_info["type"].value_counts()
print(dea_expr["pvalue"].describe())
print(dea_expr["adj_pvalue"].describe())

print("Raw significant genes:",
      (dea_expr["pvalue"] < 0.05).sum())

print("Adjusted significant genes:",
      (dea_expr["adj_pvalue"] < 0.05).sum())
sig_genes = dea_expr[
    (dea_expr["adj_pvalue"] < 0.05) &
    (abs(dea_expr["logFC"]) > 1)
]

print("Significant genes:", len(sig_genes))
top_up = sig_genes.sort_values(
    "logFC",
    ascending=False
).head(20)

top_down = sig_genes.sort_values(
    "logFC",
    ascending=True
).head(20)

print(top_up[["gene","logFC","adj_pvalue"]])
print(top_down[["gene","logFC","adj_pvalue"]])
print(expr.columns[:5])
print(log_expr.shape)
```

INTERPRETATION
These genes are among the most differentially expressed genes when comparing tumor and normal tissues across multiple cancer types in TCGA.
Differential expression analysis identified 1,353 genes with FDR-adjusted p-value < 0.05 and |log₂FC| > 1. Among the most upregulated genes were EPCAM, MMP11, CLDN4, KRT17, and CD24, which are associated with epithelial identity, extracellular matrix remodeling, and tumor progression. Several highly downregulated genes, including SOX10, TYR, TYRP1, and MLANA, are involved in melanocyte differentiation, suggesting that tissue-of-origin effects contribute substantially to transcriptomic variation in the pan-cancer cohort.




Enrichment analysis (GO and KEGG)

```python
mask = dea_expr['pvalue'].notna()
adj_pvals = np.full(len(dea_expr), np.nan)
adj_pvals[mask] = multipletests(dea_expr.loc[mask, 'pvalue'], method='fdr_bh')[1]
dea_expr['adj_pvalue'] = adj_pvals
dea_expr['clean_gene'] = dea_expr['gene'].str.split('|').str[-1]
sig_genes = dea_expr[(dea_expr["adj_pvalue"] < 0.05) & (dea_expr["logFC"].abs() > 1)]['clean_gene'].tolist()
print(f"Success! Found {len(sig_genes)} significant genes.")
print("Sample of cleaned IDs:", sig_genes[:5])
if len(sig_genes) > 0:
    import gseapy as gp
    enr = gp.enrichr(gene_list=sig_genes,
                     gene_sets=['GO_Biological_Process_2023', 'KEGG_2021_Human'],
                     organism='human', # Entrez IDs work with 'human'
                     outdir=None)
    print("Enrichment Analysis Complete!")
else:
    print("Still 0 genes? Check if any 'adj_pvalue' are actually < 0.05.")
```

```python
import mygene
import gseapy as gp

mg = mygene.MyGeneInfo()
conversion_results = mg.querymany(sig_genes, 
                                  scopes='entrezgene', 
                                  fields='symbol', 
                                  species='human',
                                  as_dataframe=True)

if 'symbol' in conversion_results.columns:
    sig_symbols = conversion_results['symbol'].dropna().unique().tolist()
    print(f"Success! Translated into {len(sig_symbols)} Gene Symbols.")
    print("Example symbols:", sig_symbols[:10])
else:
    print("Conversion failed. Check if your IDs are actually human Entrez IDs.")
    sig_symbols = []
if len(sig_symbols) > 0:
    enr = gp.enrichr(gene_list=sig_symbols,
                     gene_sets=['GO_Biological_Process_2023', 'KEGG_2021_Human'],
                     organism='human',
                     outdir=None)
    
    print(f"Enrichment done! Total results found: {len(enr.results)}")
```

```python
import matplotlib.pyplot as plt
from gseapy import barplot, dotplot

kegg_results = enr.results[enr.results['Gene_set'] == 'KEGG_2021_Human']
if not kegg_results.empty:
    barplot(kegg_results, column="Adjusted P-value", group='Gene_set', 
            size=10, top_term=10, figsize=(6, 8), color='salmon', cutoff=1.0)
    plt.title("Top 10 KEGG Pathways")
    plt.show()
go_results = enr.results[enr.results['Gene_set'] == 'GO_Biological_Process_2023']
if not go_results.empty:
    dotplot(go_results, column="Adjusted P-value", x='Gene_set', 
            size=10, top_term=10, figsize=(7, 10), title="Top 10 GO Biological Processes")
    plt.show()
```

INTERPRETATION
athway enrichment analysis revealed strong overrepresentation of ECM-receptor interaction, focal adhesion, and PI3K-Akt signaling pathways. 

Gene Ontology analysis further highlighted extracellular matrix organization and collagen fibril organization as dominant biological processes. 

Together, these findings suggest that alterations in the tumor microenvironment, cell adhesion, and proliferative signaling represent major transcriptional programs shared across diverse cancer types.
